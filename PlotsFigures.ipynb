```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def paraboloid(x, y):
    ''' z = x**2 + y**2 '''
    return (x**2 + y**2)

def plane(x,y,coeff):
    '''ax+by+cz=d
       z = Ax+By+C
    '''
    A = coeff[1]
    B = coeff[2]
    C = coeff[0]
    return A*x+B*y+C
    
xLow = - 10
yLow = xLow
xHigh = -xLow
yHigh = -yLow
N = 100
x = np.linspace(xLow, xHigh, N)
y = np.linspace(yLow, yHigh, N)
X, Y = np.meshgrid(x, y)
Z = paraboloid(X, Y)
Z = plane(X,Y,[-1,-1,0])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
ax.view_init(60, 35)
fig
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe');
```

```python
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = paraboloid(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none');
```
