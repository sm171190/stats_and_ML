```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import random
```

```python
def sampleStdNormal(N):
    sample = np.random.randn(N)
    return np.mean(sample)
   
NSamplings=  1000
meansperSample = []
for i in range(NSamplings):    
    sampleSize = 50
    meansperSample.append(sampleStdNormal(sampleSize))

print(f"Mean of Means: {np.mean(meansperSample)}")
print(f"Variance of Means: {np.var(meansperSample)}")
print(f"Median of Means: {np.median(meansperSample)}")


plt.hist(meansperSample)
plt.title("Histogram")
plt.show()
plt.boxplot(meansperSample)
plt.title("BoxPlot")
plt.show()
sns.distplot(meansperSample).set_title("DIstribution")

```

```python
def sampleExponential(mean,N):
    sample = np.random.exponential(mean,(N,1))
    return np.mean(sample)
   
NSamplings=  1000
meansperSample = []
for i in range(NSamplings):    
    sampleSize = 50
    meansperSample.append(sampleExponential(1,sampleSize))

print(f"Mean of Means: {np.mean(meansperSample)}")
print(f"Variance of Means: {np.var(meansperSample)}")
print(f"Median of Means: {np.median(meansperSample)}")


plt.hist(meansperSample)
plt.title("Histogram")
plt.show()
plt.boxplot(meansperSample)
plt.title("BoxPlot")
plt.show()
sns.distplot(meansperSample).set_title("DIstribution")
plt.show()
```

```python
def simulateGamev1(W,B):    
    N = W+B
    if N==1:
        return (W,B)
    else:
        if W>1 and B>1:
            availableChoices = ['W','B']
            B1 = random.choice(availableChoices)
            B2=random.choice(availableChoices)                        
        
        if W==1 and B>1:
            availableChoices = ['W','B']
            B1 = random.choice(['W','B'])
            if B1 == 'W':
                B2 = 'B'
            else:
                B2=random.choice(['W','B'])
            
        if W>1 and B==1:
            availableChoices = ['W','B']
            B1 = random.choice(['W','B'])
            if B1 == 'B':
                B2 = 'W'
            else:
                B2=random.choice(['W','B'])
        
        if W==0 and B>1:
            B1='B'
            B2 = 'B'
            
        if B==0 and W>1:
            B1='W'
            B2 = 'W'
        
        if B==1 and W==1:            
            B1 = random.choice(['W','B'])
            if B1 == 'B':
                B2 = 'W'
            else:
                B2='B'
        
        if B1+B2 == 'WW':
                W-=1
        elif B1+B2 == 'BB':
                B-=2
                W+=1
        else:
             W-=1
        
        return simulateGamev1(W,B)
            
```

```python
def simulateParty(N):    
    duplicates = {}    
    for i in range(1,N):    
        day = random.randint(1,365)        
        if day in list(duplicates.keys()):
            duplicates[day]+=1
        else:
            duplicates[day]=1
    
    return max(duplicates.values())

NPeople = list(range(2,100))
Nsimulations = 1000
probVsN = []

for N in NPeople:
    countVsN = []

    for simulationNo in range(Nsimulations):
        result = simulateParty(N)
        countVsN.append(result>1)
        
#     print(countVsN)
    probVsN.append(countVsN.count(True)/Nsimulations)
        
for i in range(len(probVsN)):
    if probVsN[i]>0.5:
        break


plt.plot(NPeople,probVsN)
plt.title(f"{Nsimulations} simulations per N")
plt.show()    

print(f"50% probability occurs first at N = {i} people")
```
